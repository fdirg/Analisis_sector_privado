```python
# Convertimos la columna de fecha de las 3 tablas al formato datetime

df_puestos['indice_tiempo'] = pd.to_datetime(df_puestos['indice_tiempo'])
df_sueldo_bruto['indice_tiempo'] = pd.to_datetime(df_sueldo_bruto['indice_tiempo'])
df_sueldo_neto['indice_tiempo'] = pd.to_datetime(df_sueldo_neto['indice_tiempo'])
```

```python
# Borramos la columna 'sin_clasificar' y 'total_sector_privado' de los 3 df
# Eliminamos la columna 'total_sector_privado' ya que, al ser una sumatoria de todos los puestos de cada provincia, despues podemos crearla en una dimension

borrar_columna = ['total_sector_privado', 'sin_clasificar']

df_puestos = df_puestos.drop(columns=borrar_columna)
df_sueldo_bruto = df_sueldo_bruto.drop(columns=borrar_columna)
df_sueldo_neto = df_sueldo_neto.drop(columns=borrar_columna)
```

```python
# Convertimos las tablas a una formato largo, creando una columna para todas las provincias, y otra columna para los puestos de trabajo

df_puestos = df_puestos.melt(id_vars=['indice_tiempo'], var_name='Provincias', value_name='Puestos')
df_sueldo_bruto = df_sueldo_bruto.melt(id_vars=['indice_tiempo'], var_name='Provincias', value_name='Sueldo_bruto')
df_sueldo_neto = df_sueldo_neto.melt(id_vars=['indice_tiempo'], var_name='Provincias', value_name='Sueldo_neto')
```

```python
# Verificamos el tamaño de cada Dataframe

print(f"Dimensiones del DataFrame de puestos: {df_puestos.shape}")
print(f"Dimensiones del DataFrame de sueldos brutos: {df_sueldo_bruto.shape}")
print(f"Dimensiones del DataFrame de sueldos netos: {df_sueldo_neto.shape}")
```

```python
# Unificamos las 3 tablas en una mediante la funcion merge

df_temp = pd.merge(df_puestos, df_sueldo_bruto, on=['indice_tiempo', 'Provincias'])

df = pd.merge(df_temp, df_sueldo_neto, on=['indice_tiempo', 'Provincias'])
```

```python
# Mostramos la informacion de nuestro df unico

print(df.head())

print("\nInformación del DataFrame:")
df.info()

print("\nEstadísticas Descriptivas del DataFrame:")
print(df.describe())

print("\nDimensiones del DataFrame:", df.shape)
```

```python
# Cambiamos el nombre de la columna 'indice_tiempo' a 'Fecha'

df = df.rename(columns={'indice_tiempo': 'Fecha'})
```

```python
# Ordenamos el df en base a la fecha y provincia

df = df.sort_values(by=['Fecha', 'Provincias'])

df.head()
```
